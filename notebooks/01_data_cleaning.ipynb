```python
# Catches ALL common missing value disguises at once
missing_disguises = ['?', 'NA', 'N/A', 'na', 'n/a', 
                     'none', 'None', 'NONE', '-', 
                     '--', 'unknown', 'Unknown', ' ', '']

for col in df.columns:
    for disguise in missing_disguises:
        count = (df[col] == disguise).sum()
        if count > 0:
            print(f"{col} → '{disguise}' appears {count} times")
```

```python
# Values that look empty but aren't
for col in df.select_dtypes(include='object').columns:
    whitespace = df[col].str.strip().eq('').sum()
    if whitespace > 0:
        print(f"{col} has {whitespace} whitespace-only values")
```

# Cleaning the data

```python
# Columns to drop

cols_to_drop = [
    'encounter_id',    # ID column
    'patient_nbr',     # ID column
    'examide',         # Zero variance
    'citoglipton',     # Zero variance
    'weight',          # 96.8% missing
    'payer_code',      # 39.6% missing
    'medical_specialty' # 49.1% missing
]

df.drop(columns=cols_to_drop, inplace=True)
print("Remaining columns:", df.shape[1])
```

```python
# Impute missing values with mode

cols_to_impute = ['race', 'diag_1', 'diag_2', 'diag_3']

for col in cols_to_impute:
    mode_value = df[col].mode()[0]
    df[col] = df[col].fillna(mode_value)

print("Missing values remaining:")
print(df.isnull().sum().sum())
```

```python
# Convert target to binary

df['readmitted'] = (df['readmitted'] == '<30').astype(int)

print("Target distribution after conversion:")
print(df['readmitted'].value_counts())
```

```python
df.info()
```

```python
# Restore "None" as a valid category for test result columns
df['max_glu_serum'] = df['max_glu_serum'].fillna('None')
df['A1Cresult'] = df['A1Cresult'].fillna('None')

# Verify no missing values remain
print("Missing values remaining:")
print(df.isnull().sum().sum())
```

```python
# Final Check

print("Dataset Shape:", df.shape)
print("\nMissing Values:", df.isnull().sum().sum())
print("\nTarget Distribution:")
print(df['readmitted'].value_counts())
```

```python
# Save cleaned data to processed folder

df.to_csv('../data/processed/cleaned_data.csv', index=False)
print("Cleaned data saved successfully!")
```
